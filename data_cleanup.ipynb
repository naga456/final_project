# Data Prep


```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Wuzzuf_Job_Posts_Sample.csv')
df = data[['job_category1','job_description']]
df=df.rename(columns = {'job_category1':'label'})
df = df.dropna()
```

```python
df['label'].value_counts()
```

```python
categories = ['IT/Software Development', 'Engineering','Customer Service/Support','Marketing','Sales/Retail/Business Development']
print(categories)
```

```python
df[df['label'].isin(categories)]
```

```python
#wl - Clean out the HTML tags first
import re
df.job_description = [re.sub(r'[<][^<>]*[>]', '', x) for x in df.job_description]
print("new corpus==", df.job_description)
```

```python
# create a .csv of the database to import it into the elephantsql database
df.to_csv("clean_data.csv", index=False)
```

# Create model

```python
X = df['job_description']
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

#wl-Remove stop words
model = make_pipeline(TfidfVectorizer(stop_words='english'), MultinomialNB())
```

```python
model.fit(X_train,y_train)
```

```python
#wl - Save the model
>>> from joblib import dump, load
>>> dump(model, 'nb.joblib') 
```

```python
#wl - score the model
model.score(X_test,y_test)
```

```python
labels = model.predict(X_test)
```
